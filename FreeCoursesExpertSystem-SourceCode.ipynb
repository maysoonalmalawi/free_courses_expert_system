```python
from experta import *


# -------------------preparing course fact and the knoweldge base
class Course(Fact):
  pass

class CourseKBA(KnowledgeEngine):
  
  # 1- science rule
  @Rule(Course(sci_ans=('y')))
  def guess_science(self):
    print("- Natural Sciences \n- Astronomy")

  # 2- art rule
  @Rule(Course(art_ans=('y')))
  def guess_arts(self):
    print("- Fashion")

  # 3- health rule
  @Rule(Course(health_ans=('y')))
  def guess_health(self):
    print("- Nutrition \n- Personal Health")

  # 4- religions rule
  @Rule(Course(religion_ans=('y')))
  def guess_religion(self):
    print("- Quraan Studies \n- Prophetic Biography (Sirah) \n- Prohetic Biography for Children")

  # 5- social sciences rule
  @Rule(Course(social_ans=('y')))
  def guess_social(self):
    print("- Principles of Political Sciences \n- Educational System in Saudi Arabia \n- Tourist Texts")

  # 6- media and communication rule
  @Rule(Course(media_ans=('y')))
  def guess_media(self):
    print("- Introduction to the Media \n- Persuasion \n- English for Communication \n- Photojournalism")

  # 7- business rule
  @Rule(Course(business_ans=('y')))
  def guess_business(self):
    print("- Public Relations \n- Personal Project Management")

  # 8- languages rule
  @Rule(Course(lang_ans=('y')))
  def guess_lang(self):
    print("- French \n- English")


  # 9- technology rule
  @Rule(Course(tech_ans=('y')))
  def guess_tech(self):
    print("- Artificial Intelligence")

  #10- psychology rule
  @Rule(Course(psych_ans=('y')))
  def guess_psych(self):
    print("- General Psychology \n- Introduction to Psychology \n- Criminal Psychology")


# -------------------preparing the engine
engine = CourseKBA()

engine.reset()

# -------------------printing the welcome message

print("                                     Welcome to the Free Courses Expert System!!!")
print("---------------------------------------------------------------------------------------------------------------------")
print("By the end of this survey, you will hopefully be able to have a clear view of what free courses you should take!")
print("Please answer each of the ten following questions with y (for yes) or n (for no)")
print("---------------------------------------------------------------------------------------------------------------------\n")


# -------------------printing the questions

# 1- science question
sci_answer = input("1- Are you interested in topics like chemistry, physics, astronomy, zoology and the human body?(y|n) ").lower()

# 2- arts question
art_answer = input("2- Are you into art and design and enjoy activities like hand embroidery and learning about textile chemistry?(y|n) ").lower()

# 3- health question
health_answer = input("3- Do you try to lead a healthy lifestyle and are interested in general health and balanced food specifications?(y|n) ").lower()

# 4- religion question
religion_answer = input("4- Are you interested in Islamic studies, Islamic law, and Tajweed?(y|n) ").lower()

# 5- social sciences question
social_answer = input("5- Are you interested in political thought or the relationships between countries?(y|n) ").lower()

# 6- media question
media_answer = input("6- Are you interested in photography, journalism, or media?(y|n) ").lower()

# 7- business question
business_answer = input("7- Are you interested in interacting and communicating in the work environment?(y|n) ").lower()

# 8- languages question
languages_answer = input("8- Are you interested in literature, foreign languages, or linguistics?(y|n) ").lower()

# 9- technology question
tech_answer = input("9- Are you interested in computer science, technology, or programming?(y|n)? ").lower()

# 10- psychology question
psych_answer = input("10- Do you enjoy reading and learning about psychology, mental disorders, and the theories of personality?(y|n) ").lower()



# -------------------declaring the engine
engine.declare(Course(sci_ans = sci_answer),
               Course(art_ans = art_answer), 
               Course(health_ans = health_answer), 
               Course(religion_ans = religion_answer),
               Course(social_ans = social_answer),
               Course(media_ans = media_answer),
               Course(business_ans = business_answer),
               Course(languages_ans = languages_answer),
               Course(tech_ans = tech_answer),
               Course(psych_ans = psych_answer)
               )

# -------------------printing the output 
print("\n---------------------------------------------------------------------------------------------------------------------\n")
print("The suitable free course for you are: ")

# -------------------run the engine
engine.run()

# -------------------print closing message
print("\n---------------------------------------------------------------------------------------------------------------------")
print ("                                         Thank you for using our system!!!")    
```
